```python
import pandas as pd
```

```python
df = pd.read_csv("../data/stops.csv")
df
```

```python
clean_df = df.copy()
```

```python
color_mapping = {
    "A": "#0039a6",
    "C": "#0039a6",
    "E": "#0039a6",
    "B": "#ff6319",
    "D": "#ff6319",
    "F": "#ff6319",
    "M": "#ff6319",
    "G": "#6cbe45",
    "L": "#a7a9ac",
    "J": "#996633",
    "Z": "#996633",
    "N": "#fccc0a",
    "Q": "#fccc0a",
    "R": "#fccc0a",
    "W": "#fccc0a",
    "1": "#ee352e",
    "2": "#ee352e",
    "3": "#ee352e",
    "4": "#00933c",
    "5": "#00933c",
    "6": "#00933c",
    "7": "#b933ad",
    "T": "#00add0",
    "S": "#808183",
}
```

```python
divison_mapping = {
    "A": "B",
    "C": "B",
    "E": "B",
    "B": "B",
    "D": "B",
    "F": "B",
    "M": "B",
    "G": "B",
    "L": "B",
    "J": "B",
    "Z": "B",
    "N": "B",
    "Q": "B",
    "R": "B",
    "W": "B",
    "1": "A",
    "2": "A",
    "3": "A",
    "4": "A",
    "5": "A",
    "6": "A",
    "7": "A",
    "S": "S"
}
```

```python
clean_df["temp"] = clean_df["the_geom"].apply(lambda x: x.split(" "))
clean_df["lat"] = clean_df["temp"].apply(lambda x: x[1].replace("(", ""))
clean_df["long"] = clean_df["temp"].apply(lambda x: x[2].replace(")", ""))
clean_df["color"] = clean_df["line"].apply(lambda x: color_mapping[x[0]])
clean_df["division"] = clean_df["line"].apply(lambda x: divison_mapping[x[0]])

clean_df.columns = clean_df.columns.str.lower()
clean_df[["name", "line", "notes", "lat", "long", "color", "division"]].to_csv("../data/stops.csv", index=False)
```
